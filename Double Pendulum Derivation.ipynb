```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import solve_ivp
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

Set variables

```python
x1, x2, y1, y2, L1, L2, m1, m2, t, g = smp.symbols(r'x_1 x_2 y_1 y_2 L_1 L_2 m_1 m_2 t g', real = True)
```

Set $\theta_1$ and $\theta_2$ as functions of time, $t$

```python
ftheta1 = smp.Function(r'\theta_1')(t)
ftheta2 = smp.Function(r'\theta_2')(t)
```

Define $x_1, x_2, y_1, y_2$ in terms of $L_1, L_2,$ and theta function $(\theta_1(t), \theta_2(t))$

```python
x1 = L1 * smp.sin(ftheta1)
x2 = L1 * smp.sin(ftheta1) + L2 * smp.sin(ftheta2)
y1 = -L1 * smp.cos(ftheta1)
y2 = -L1 * smp.cos(ftheta1) -L2 * smp.cos(ftheta2)
```

Differentiate!

```python
x1dot = smp.diff(x1, t)
x2dot = smp.diff(x2, t)
y1dot = smp.diff(y1, t)
y2dot = smp.diff(y2, t)
```

```python
y2dot
```

Define $T$ in $\mathcal{L} = T - V$

```python
T1 = (1/2) * m1 * (((x1dot)**2) + ((y1dot)**2))
T2 = (1/2) * m2 * (((x2dot)**2) + ((y2dot)**2))
T = T1 + T2
```

Define $V$ in $\mathcal{L} = T - V$

```python
V1 = m1 * g * y1
V2 = m2 * g * y2
V = V1 + V2
```

LAGRANGIAN!!

```python
L = T-V
```

```python
L = L.simplify()
L
```

The Euler-Lagrange equation: $\Huge\frac{\partial \mathcal{L}}{\partial \theta} - \frac{d}{dt} \frac{\partial \mathcal{L}}{\partial \dot{\theta}} = 0$

```python
#First we have to define theta dots^^ hehe
theta1dot = smp.diff(ftheta1, t)
theta2dot = smp.diff(ftheta2, t)
theta1ddot = smp.diff(ftheta1, t, t)
theta2ddot = smp.diff(ftheta2, t, t)
```

```python
LE1 = smp.diff(L, ftheta1) - smp.diff(smp.diff(L, theta1dot), t)
LE2 = smp.diff(L, ftheta2) - smp.diff(smp.diff(L, theta2dot), t)
```

Lagrange Equations

```python
LE1 = LE1.simplify()
LE1
```

```python
LE2 = LE2.simplify()
LE2
```

Solve for $\ddot{\theta_1}$ and $\ddot{\theta_2}$  

```python
sol = smp.solve((LE1, LE2), (theta1ddot, theta2ddot), dict=True)[0]

theta1ddot_sol_eq = sol[theta1ddot]
theta2ddot_sol_eq = sol[theta2ddot]
```

The Equations!

```python
theta1ddot_sol_eq
```

```python
theta2ddot_sol_eq.simplify()
```

```python
calc_alpha_1 = smp.lambdify((L1, L2, m1, m2, g, ftheta1, ftheta2, theta1dot, theta2dot), theta1ddot_sol_eq )
calc_alpha_2 = smp.lambdify((L1, L2, m1, m2, g, ftheta1, ftheta2, theta1dot, theta2dot), theta2ddot_sol_eq )
```

```python
calc_alpha_1
```

```python
m1_val = 1
m2_val = 1
g_val = 9.81
L1_val = 5
L2_val = 1
theta_1_0 = 10 * np.pi/180    
dthedt_1_0 = 1
theta_2_0 = np.pi
dthedt_2_0 = 10
```

```python
def dSdt(t, S):
    theta1_val, omega1_val, theta2_val, omega2_val = S

    a1 = calc_alpha_1(L1_val, L2_val, m1_val, m2_val, g_val, theta1_val, theta2_val, omega1_val, omega2_val)
    a2 = calc_alpha_2(L1_val, L2_val, m1_val, m2_val, g_val, theta1_val, theta2_val, omega1_val, omega2_val)

    return [omega1_val, a1, omega2_val, a2]
```

```python
t_eval = np.linspace(0, 20, 2400)

y0 = [theta_1_0, dthedt_1_0, theta_2_0, dthedt_2_0]

sol = solve_ivp(
    dSdt,
    (0, 20),
    y0,
    t_eval = t_eval,
    method='DOP853',     
    rtol=1e-12,
    atol=1e-12,
    max_step=0.005       
)

theta1_vals = sol.y[0]
omega1_vals = sol.y[1]
theta2_vals = sol.y[2]
omega2_vals = sol.y[3]
```

```python
x1_vals = L1_val * np.sin(theta1_vals)
y1_vals = -L1_val * np.cos(theta1_vals)

x2_vals = x1_vals + L2_val * np.sin(theta2_vals)
y2_vals = y1_vals - L2_val * np.cos(theta2_vals)
```

```python
import matplotlib as mpl
mpl.rcParams['animation.embed_limit'] = 1000
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML


fps = 24

duration = sol.t[-1] - sol.t[0]          
target_frames = int(round(duration * fps))  

gif_name = "double_pendulum_professional.gif"
tail_length = 50

frame_idx = np.linspace(0, len(sol.t) - 1, target_frames, dtype=int)

t_anim  = sol.t[frame_idx]
x1_anim = x1_vals[frame_idx]
y1_anim = y1_vals[frame_idx]
x2_anim = x2_vals[frame_idx]
y2_anim = y2_vals[frame_idx]


fig, ax = plt.subplots(figsize=(6.2, 6.2), dpi=100)

total_length = L1_val + L2_val + 0.25
ax.set_xlim(-total_length, total_length)
ax.set_ylim(-total_length, total_length)
ax.set_aspect("equal")

ax.set_title("Double Pendulum", fontsize=14, pad=10)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, color="#d9d9d9", linewidth=0.8, alpha=0.7)
ax.set_facecolor("white")


rod, = ax.plot([], [], '-', lw=2.0, color="#2f2f2f")
bob1, = ax.plot([], [], 'o', ms=7, color="#4c78a8")
bob2, = ax.plot([], [], 'o', ms=8, color="#e45756")

trace_full, = ax.plot([], [], '-', lw=1.0, color="#a0a0a0", alpha=0.25)
trace_tail, = ax.plot([], [], '-', lw=1.6, color="#4c78a8", alpha=0.9)

time_text = ax.text(
    0.03, 0.96, "",
    transform=ax.transAxes,
    ha="left", va="top",
    fontsize=10,
    color="#333333"
)


def init():
    rod.set_data([], [])
    bob1.set_data([], [])
    bob2.set_data([], [])
    trace_full.set_data([], [])
    trace_tail.set_data([], [])
    time_text.set_text("")
    return rod, bob1, bob2, trace_full, trace_tail, time_text


def update(j):
    rod.set_data(
        [0, x1_anim[j], x2_anim[j]],
        [0, y1_anim[j], y2_anim[j]]
    )

    bob1.set_data([x1_anim[j]], [y1_anim[j]])
    bob2.set_data([x2_anim[j]], [y2_anim[j]])

    trace_full.set_data(x2_anim[:j+1], y2_anim[:j+1])

    start = max(0, j - tail_length)
    trace_tail.set_data(x2_anim[start:j+1], y2_anim[start:j+1])

    time_text.set_text(f"t = {t_anim[j]:.2f} s")
    return rod, bob1, bob2, trace_full, trace_tail, time_text


ani = FuncAnimation(
    fig,
    update,
    frames=len(t_anim),
    init_func=init,
    interval=1000 / fps,
    blit=True,
    repeat=False
)

HTML(ani.to_jshtml(fps=fps))
```
